```python
import pandas as pd
import numpy as np

df = pd.read_csv('myexcel - myexcel.csv.csv')
```

```python
df.head
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Height'] = df['Height'].apply(lambda x: np.random.randint(150, 181) if pd.isna(x) or not isinstance(x, (int, float)) else x)
```

```python
df['Height'].describe()
```

```python
team_distribution = df['Team'].value_counts()

team_percentage = (team_distribution / len(df)) * 100
team_summary = pd.DataFrame({
    'Employee Count': team_distribution,
    'Percentage': team_percentage
})
team_summary
```

```python
position_distribution = df['Position'].value_counts()

position_distribution
```

```python
bins = [20, 30, 40, 50, 60, 100]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Get the count of employees in each age group
age_group_distribution = df['age_group'].value_counts()

age_group_distribution
```

```python
salary_by_team_position = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

max_salary_team_position = salary_by_team_position.loc[salary_by_team_position['Salary'].idxmax()]

max_salary_team_position
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
correlation = df['Age'].corr(df['Salary'])
correlation
```

```python
team_distribution.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
position_distribution.plot(kind='bar', color='red')
plt.title('Employee Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

```python
age_group_distribution.plot(kind='bar', color='green')
plt.title('Employee Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
sns.barplot(x='Salary', y='Team', data=salary_by_team_position, hue='Position', dodge=False)
plt.title('Salary Expenditure by Team and Position')
plt.xlabel('Total Salary Expenditure')
plt.ylabel('Team')
plt.show()
```
